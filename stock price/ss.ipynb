```python
import pandas as pd 
```

```python
df=pd.read_csv(r"D:\work\natural language processing\stock price\r", encoding='ISO-8859-1')
df.head(2)
```

```python
train = df[df['Date'] < '20150101']# here i have selected the test data only uoto these row 
test = df[df['Date'] > '20141231'] # this is the test data
```

here i b doing some data pre processing in which i will use regular exression library

```python
# removing the panchuaction marks 
data = train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ", regex=True, inplace= True)

# removing the column name for easily intrupption 
list1=[i for i in range(25)]
new_index=[str(i) for i in list1]
data.columns=new_index
data.head(2)
```

# FOR CONVERTING THE DATA INTO LOWER FORM 

```python
# converting the data into lower form
for index in new_index:
    data[index]=data[index].str.lower()
data.head(4)
```

JOINING ALL WORD TO A PARAGRAPH SO THAT WE CAN APPLY THE WORD OF BAG OR TFIDF

```python
headline=[]# making empty list for storing the paragraph
for row in range (0,len(data.index)):# runninng the loop from 0 to length of data indexed
    headline.append(" ".join(str(x)for x in data.iloc[row,0:25]))
```

```python
headline[1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer# importing the bag of words model for vectoriziation 
from sklearn.ensemble import RandomForestClassifier# importing the random forest for the classfication 
```

```python
# importing the bag of words 
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headline)
```

```python
# imppolementing randomforest classifier
rd=RandomForestClassifier(n_estimators=200,criterion="entropy")
rd.fit(traindataset,train['Label'])
```

```python
# predictind for the test dataset
test_transform=[]
for row in range (0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset=countvector.transform(test_transform)
predictions= rd.predict(test_dataset)
```

```python
# immporting the library to cheak the accuracy of the model
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
